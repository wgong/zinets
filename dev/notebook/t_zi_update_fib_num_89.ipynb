```python
file_csv = "fib_num_89-144-377-updates.csv"
```

```python
df = pd.read_csv(file_csv).fillna("")
```

```python
data = df.to_dict("record")
```

```python
for d in data: 
    zi = d.get("zi")
    fib_num  = d.get("fib_num")
    nstrokes = d.get("nstrokes", "")
    print(zi, fib_num, int(nstrokes))
```

## Update fib_num

```python
with DBConn(Path(file_db)) as _conn:
    cur = _conn.cursor()
    for d in data: 
        zi = d.get("zi")
        fib_num  = str(d.get("fib_num"))
        nstrokes = d.get("nstrokes", "")
        if nstrokes:
            nstrokes = str(int(nstrokes))
        upd_sql = f"""
            update t_zi set 
                fib_num = '{fib_num}'
            where zi = '{zi}';
        """
        print(upd_sql)
        cur.execute(upd_sql)
        
    _conn.commit()
```

## Update nstrokes

```python
with DBConn(Path(file_db)) as _conn:
    cur = _conn.cursor()
    for d in data: 
        zi = d.get("zi")
        fib_num  = str(d.get("fib_num"))
        nstrokes = d.get("nstrokes", "")
        if not nstrokes: continue 
            
        nstrokes = str(int(nstrokes))
        upd_sql = f"""
            update t_zi set 
                nstrokes = '{nstrokes}'
            where zi = '{zi}' and nstrokes is null
        """
        print(upd_sql)
        cur.execute(upd_sql)
        
    _conn.commit()
```
