```python
URL = "https://unicode.org/emoji/charts/full-emoji-list.html"
```

```python
with open("Full Emoji List, v15.1.html", encoding="utf-8") as f:
    txt = f.read()
```

```python
soup = BeautifulSoup(txt, "html.parser")
```

```python
results = soup.find("tbody").find_all("tr")
```

```python
data = []
for row in results:
    tds = row.find_all("td")
    if len(tds) != 9: continue
    try:
        id_ = tds[0].text 
        unicode_ = tds[1].text
        browser_ = tds[2].text
        sample_ = tds[3].text
        name_ = tds[-1].text 
        data.append([id_, unicode_, browser_, sample_, name_])
    except Exception as ex:
        print(f"[ERR] {row}")
```
