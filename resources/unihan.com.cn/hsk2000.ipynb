```python
BASE_URL = "https://hanzi.unihan.com.cn"
```

```python
urls = []

# # 汉语水平考试汉字列表
urls += [f"https://hanzi.unihan.com.cn/School/HSK/{i}" for i in range(1,7)]

## Common 
urls += [f"https://hanzi.unihan.com.cn/School/Common/{i}" for i in range(1,17)]

## Less Common 
urls += [f"https://hanzi.unihan.com.cn/School/CommonLow/{i}" for i in range(1,8)]
```

```python
# going to scrap each page to get each Zi and its URL
```

```python
urls[0].replace("https://hanzi.unihan.com.cn/School/", "").replace("/","-")
```

```python
def parse_td(td):
    try:
        zi = td.text.strip()
        url = td.find("a")['href']
        return zi, f"{BASE_URL}{url}"
    except:
        return "", ""
```

```python
def parse_page(url_1):
    category = url_1.replace("https://hanzi.unihan.com.cn/School/", "").replace("/","-")
    data = []
    page = requests.get(url_1, headers=BROWSER_HEADERS)
    soup = BeautifulSoup(page.content, "html.parser")
    tbody = soup.find("tbody")
    tds = tbody.find_all("td")
    for td in tds:
        zi, url = parse_td(td)
        if zi and url:
            data.append([zi, url, category])
    return  data
```

```python
data = []
for url_1 in urls:
    tmp = parse_page(url_1)
    if tmp: 
        data += tmp
```
