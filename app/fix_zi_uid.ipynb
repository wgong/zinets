## reset u_id of t_zi

```python
sql_stmt = f"""
select max(cast(u_id as int)) max_uid from t_zi
"""

with DBConn() as _conn:
    df1 = pd.read_sql(sql_stmt, _conn).fillna("")
```

```python
max_uid = df1["max_uid"][0]
```

```python
for zi in df["zi"].to_list():
    max_uid += 1
    u_id = str(max_uid)
    sql_stmt = f"""
        update t_zi set u_id = '{u_id}' where zi = '{zi}' and u_id is null
    """
    print(sql_stmt)
    with DBConn() as _conn:
        db_run_sql(sql_stmt, _conn)
```

## reset u_id of t_zi_part

```python
sql_stmt = f"""
select z.u_id as z_uid, zp.u_id as zp_uid, z.* 
from t_zi z, t_zi_part zp where z.zi = zp.zi and z.u_id != zp.u_id;
"""

with DBConn() as _conn:
    df2 = pd.read_sql(sql_stmt, _conn).fillna("")
```

```python
df2 
```

```python
for zi in df2["zi"].to_list():
    sql_stmt = f"""
        update t_zi_part 
        set u_id = (
            select u_id from t_zi  
            where zi = t_zi_part.zi
            limit 1
        ) 
        where zi = '{zi}'
    """
    print(sql_stmt)
    with DBConn() as _conn:
        db_run_sql(sql_stmt, _conn)
```
