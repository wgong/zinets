```python
with DBConn() as _conn:
    sql_0 = """
    select trim(zi)  as zi
 from t_ele_zi
 where is_active='Y' and length(trim(zi)) > 0
 union 
     select trim(variant)  as zi
 from t_ele_zi
 where is_active='Y' and length(trim(variant)) > 0
    """
    
    df_ele_zi = pd.read_sql(sql_0, _conn).fillna("")

    ele_zi_set = set()
    for i in df_ele_zi["zi"].to_list():
        i = i.strip()
        if not i: continue
        
        if len(i)> 1:
            for c in i:
                ele_zi_set.add(c)
        else: 
            ele_zi_set.add(i)     
```

```python
len(ele_zi_set), ele_zi_set  # 427  # 495 after including variants  # 537 after merging some characters from shuowen book
```

```python
with DBConn() as _conn:
    sql_0 = """
    select zi
 from t_ele_zi
 where is_active='Y' and id_kangxi > 0 and length(trim(zi)) > 0
    """
    
    df_kangxi = pd.read_sql(sql_0, _conn).fillna("")
    kangxi_rad_set = set(df_kangxi["zi"].to_list())
```

```python
len(kangxi_rad_set), kangxi_rad_set   # 245
```
