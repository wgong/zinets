```python
cat = ""
zi_dic = {}
for x in [c.strip() for c in text_raw.split("\n") if c.strip()]:
    if "-" in x:
        cat = x.replace("- ", "").strip()
    else:
        zi_dic[cat] = zi_dic.get(cat,"") + x.replace(".","").replace("。","").replace(",","").replace("，","").replace("(","").replace(")","")
```

```python
zi_dic
```

```python
def reverse_dictionary(original_dict):
    """
    Reverses the given dictionary, making Chinese characters keys and their categories values.

    Args:
        original_dict (dict): The original dictionary with categories as keys and lists of Chinese characters as values.

    Returns:
        dict: The reversed dictionary with Chinese characters as keys and lists of categories as values.
    """

    reversed_dict = {}
    for category, characters in original_dict.items():
        for character in characters:
            if not character.strip(): 
                continue 
            if character not in reversed_dict:
                reversed_dict[character] = set()
            reversed_dict[character].add(category)
    return reversed_dict

# 示例用法
original_dict = {'Nature': '金木水火土气',
                 'Space-Time': '古今历史久气'}

reversed_dict = reverse_dictionary(original_dict)
print(reversed_dict)
```

```python
zi_category = reverse_dictionary(zi_dic)
print(zi_category)
```

```python
for k,v in zi_category.items():
    if len(v) > 1:
        print(f"k={k}, v={v}")
```
