```python
part_dict = {c["zi"].strip(): c["fib_num"]   for c in parts}
char360_dict = {c["zi"].strip(): int(c["fib_num"]) if c["fib_num"] else None  for c in chars360}

tag_dict = {c["zi"].strip(): c["tag"]   for c in parts}
```

```python
char360_dict
```

```python
fib_nums = {}
for zi in set(list(part_dict.keys()) + list(char360_dict.keys())):
    fib_1 = part_dict.get(zi, None)
    fib_2 = char360_dict.get(zi, None)
    if fib_1 and fib_2:
        if fib_1 == fib_2:
#             print("[matched] ", zi,fib_1)
            fib_nums[zi] = fib_1
        else:
            print("[unmatched] ",zi,fib_1,fib_2)
    elif fib_1:
#         print("[parts] ",zi,fib_1)
        fib_nums[zi] = fib_1
    elif fib_2:
#         print("[char360] ",zi,fib_2)
        fib_nums[zi] = fib_2
```
