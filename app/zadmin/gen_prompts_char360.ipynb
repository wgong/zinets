```python
with DBConn() as _conn:
    sql_stmt = """
        with ch360 as (
            select zi,cast(fib_num as int) as fib_num 
            from t_zi 
            where fib_num is not null     
        )
        select * from ch360 order by fib_num, zi
    """
    df = pd.read_sql(sql_stmt, _conn)
    zi_360 = df.to_dict("records")
```

```python
prompt = """
================ BATCH: {batch_id}   -- PROGRESS: {progress}  -- FIB_NUM: {fib_num} ===========


You are an expert in Chinese language, 

can you generate a holistic view on this chinese character  {zi}
in terms of the following attributes:
含义
字形
读音 
字源
常用词组
成语
例句
短故事
诗词
图片
音频 
视频 
电影
参考资料
有趣网站

(1) give the answer in Chinese 
(2) format the answer in valid json and ensure quotes are properly escaped (specifically avoid double-quotes nested in doube-quotes)
(3) whenever possible, give 5 or more examples for the following attributes:

常用词组
成语
例句
短故事
诗词
图片
音频 
视频 
电影
参考资料
有趣网站


"""
```

```python
print(prompt.format(zi="日", progress="1/1", batch_id="1", fib_num="1"))
```

```python
fib_dic = {x["zi"] : x["fib_num"]   for x in zi_360}
```

```python
fib_dic
```

```python
prompt_file = "prompts-char360.txt"
NUM_PER_ROW = 10

with open(prompt_file, "w", encoding="utf-8") as f:

    x = list(fib_dic.keys())
    batch_id = 0
    for i in range(0, len(x), NUM_PER_ROW):
        f.write(f"Working on the following {NUM_PER_ROW} Characters ..........\n")
        f.write(" ".join(x[i:i+NUM_PER_ROW]))
        n = 0
        batch_id += 1
        for zi in x[i:i+NUM_PER_ROW]:
            n += 1
            f.write(prompt.format(zi=zi, progress=f"{n}/{NUM_PER_ROW}", batch_id=batch_id, fib_num=fib_dic[zi]))
```
