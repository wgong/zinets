```python
import pandas as pd 
file_path = "../devs/sql/parts_count-20240118.csv"
```

```python
s = "index,zi_part,freqs, status, alias, category"
header = [c.strip() for c in s.split(",")]
```

```python
header
```

```python
with open(file_path, encoding="utf-8") as f:
    lines = [l.strip() for l in f.read().split("\n") if l.strip() and not l.startswith("#")]
```

```python
data = []

for l in lines:
    if not l.strip(): continue 
        
    vals = [c.strip() for c in l.split(",") if c.strip()]
    for i in range(len(header)-len(vals)):
        vals.append('')
    data.append(vals)
```

```python
data
```

```python
df = pd.DataFrame(data, columns=header)
```

```python
df
```

```python
df.to_csv("zi_parts.csv", index=False)
```
